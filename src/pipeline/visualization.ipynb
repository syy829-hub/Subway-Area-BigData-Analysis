```python
df_q1 = pd.read_csv('q1_top10_stores.csv')

fig1 = px.bar(
    df_q1, 
    y='station_name', 
    x='store_count', 
    color='store_type',
    orientation='h', # 가로 막대
    title='<b>출퇴근 유동인구 Top 10 역세권 주요 업종 비율</b>',
    labels={'station_name': '지하철역', 'store_count': '상가 수 (개)', 'store_type': '상권 업종'},
    color_discrete_sequence=px.colors.qualitative.Pastel 
)
fig1.update_layout(
    barmode='stack',
    yaxis={'categoryorder': 'total ascending'},
    font=dict(family="Malgun Gothic", size=14)
)
fig1.show()
```

```python
df_q2 = pd.read_csv('q2_morning_vs_night.csv')

df_q2_melted = df_q2.melt(
    id_vars='store_type', 
    value_vars=['morning_count', 'night_count'],
    var_name='Time', 
    value_name='Count'
)
df_q2_melted['Time'] = df_q2_melted['Time'].replace({
    'morning_count': '아침 혼잡역 (08시)', 
    'night_count': '심야 혼잡역 (22~23시)'
})

fig2 = px.bar(
    df_q2_melted, 
    y='store_type', 
    x='Count', 
    color='Time', 
    barmode='group', 
    orientation='h',
    title='<b>아침 출근 집중역 vs 심야 유흥 집중역 주요 업종 비교 (Top 10)</b>',
    labels={'store_type': '', 'Count': '상가 수 (개)', 'Time': '시간대'}, 
    color_discrete_sequence=['#4285F4', '#EA4335'] 
)
fig2.update_layout(
    height=600, 
    yaxis={'categoryorder': 'total ascending'},
    font=dict(family="Malgun Gothic", size=14),
    margin=dict(l=150, r=20, t=80, b=40),
    bargroupgap=0.15 
)

fig2.show()
```

```python
df_q3 = pd.read_csv('q3_blue_ocean.csv')

fig3 = px.bar(
    df_q3, 
    y='station_name', 
    x='people_per_store', 
    orientation='h',
    text='people_per_store', 
    title='<b>블루오션 상권 Top 5 (상가 1개당 유동인구 수)</b>',
    labels={'station_name': '지하철역', 'people_per_store': '상가 1개당 유동인구 (명)'},
    color='people_per_store',
    color_continuous_scale='Viridis'
)
fig3.update_traces(
    texttemplate='%{text:,.0f} 명', 
    textposition='outside'         
)

fig3.update_layout(
    yaxis={'categoryorder': 'total ascending'},
    font=dict(family="Malgun Gothic", size=14),
    coloraxis_showscale=False 
)
fig3.show()
```

```python
df_corr = pd.read_csv('correlation.csv')
df_corr = df_corr.dropna()
df_corr = df_corr[df_corr['total_stores'] > 0]

fig4 = px.scatter(
    df_corr,
    x='peak_traffic',
    y='total_stores',
    hover_name='station_name', 
    trendline='ols',          
    title='<b>유동인구와 상권 밀집도 간의 상관관계</b>',
    labels={
        'peak_traffic': '출퇴근 유동인구 수 (명)',
        'total_stores': '총 상가 수 (개)'
    },
    opacity=0.6, 
    color_discrete_sequence=['#2b8cbe']
)
fig4.update_layout(
    font=dict(family="Malgun Gothic", size=14),
    plot_bgcolor='white', 
    hovermode='closest'
)
fig4.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGray')
fig4.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGray')

fig4.show()
```
